```python
import pandas
data=pandas.read_csv('EastWestAirlines.csv')
data
```

```python
data.isnull().sum()
```

```python
array=data.values
x=array[:,4:7]
```

```python
from sklearn.cluster import KMeans
model = KMeans(n_clusters=10, random_state=42)
model.fit(x)
```

```python
data.info()
```

```python
# Lets see the 10 Clusters Kmeans created.
means = model.cluster_centers_
clusters = pandas.DataFrame(means, columns = ['FlightMiles', 'FlightTrans', 'DaysSinceEnroll'])
clusters
```

```python
# Let see who is in group 2
data['no']  = model.labels_   # get every record group number
cluster5 =  data[data['no'] == 5]
cluster5

```

```python
data[data['no'] == 5].to_csv('Cluster5.csv')
```
